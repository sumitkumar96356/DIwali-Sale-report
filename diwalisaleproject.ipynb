## Exploratory Data analysis

```python
ds.columns
```

```python
sx=sns.countplot(x='Gender',data=ds)
for bars in sx.containers:
    sx.bar_label(bars)
```

```python
sg=ds.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sg)
```

From the above graph we can see that most of the buyuers are females and even the purchasing power of female are greater than men

## Age

```python
sa=sns.countplot(data=ds , x='Age Group', hue= 'Gender')
for bars in sa.containers:
    sa.bar_label(bars)
```

```python
# total age vs age group
sales_age=ds.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

from the above graph we can see that most of the buyers are of age group between 26-35 yrs female

## stats

```python
#total number of orders from top 10 stats
sales_state=ds.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# total amount/sales from top 10 stats
sales_state=ds.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

*from above graphs we can see that most of the orders & total sales/ amount are from uttar pradesh , maharashtra and karnatka respectively*

## marital status

```python
ms=sns.countplot(data=ds, x='Marital_Status')
sns.set(rc={'figure.figsize':(6,1)})
for bars in ms.containers:
    ms.bar_label(bars)
```

```python
ms=ds.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='Marital_Status',y='Amount',data=ms,hue='Gender')
```

*from graphs we can see that most pf the buyers married(female) and they have high purchasing power*

## occupation

```python
ds.columns
```

```python
occ=sns.countplot(data=ds,x='Occupation')
sns.set(rc={'figure.figsize':(18,10)})
for bars in occ.containers:
    occ.bar_label(bars)
```

```python
occ_state=ds.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',data=occ_state,y= 'Amount')
```

*as we can see from the above graphs most of the persons are working in IT sector,healthcare,Aviation*

## product category

```python
ds.columns
```

```python
pc=sns.countplot(x='Product_Category',data=ds)
sns.set(rc=({'figure.figsize':(30,15)}))
for bars in pc.containers:
    pc.bar_label(bars)
```

```python
ds.columns
```

```python
pcg=ds.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Product_Category')
sns.set(rc=({'figure.figsize':(30,15)}))
sns.barplot(x='Product_Category',data=pcg,y='Amount')
```

*we can see in the above graph most of the sold products are food,clothing, footwear and electronics category*
